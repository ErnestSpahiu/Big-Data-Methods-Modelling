# Newton Rophson method

```julia
using Plots

# Define the function and its derivative
f(x) = x^2 - 4
df(x) = 2*x

function newton_raphson_visualize(x0; tol=1e-6, max_iter=100)
    x = x0
    xs = [x0]
    for i in 1:max_iter
        x_new = x - f(x) / df(x)
        push!(xs, x_new)
        if abs(x_new - x) < tol
            return x_new, xs
        end
        x = x_new
    end
    return x, xs
end
```

```julia
# Initial Guess
x0 = 3.0

# Run the method
root, xs = newton_raphson_visualize(x0)

# Display results
println("Root found at: $root")

# Visualization
plot(f, -5:0.1:5, label="f(x)", title="Newton-Raphson Method", xlabel="x", ylabel="f(x)")
scatter!(xs, f.(xs), label="Iterations", color=:red)
vline!([root], label="Root", linestyle=:dash, color=:green)
```
